```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn_extra.cluster import KMedoids
import warnings
warnings.filterwarnings('ignore')
```

```python
  data= np.array([[7,8],[9,10],[11,5],[4,9],[7,5],[2,3],[4,5]])
    # created a 2D NumPy array called "data" with 7 rows and 2 columns
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x=data[:,0],y=data[:,1],marker='o',s=100)
plt.title('Dummy img data point')
plt.show()
```

```python
centers=[[4,5],[9,10]]
```

```python
kmedoids=KMedoids(n_clusters=2)
kmedoids.fit(data)
labels=kmedoids.labels_
#  labels will be integers from 0 to 1 (since you specified 2 clusters)
```

```python
unique_labels = set(labels)
colors_plot = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
xy = data
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x=xy[:, 0], y=xy[:, 1], marker='+', hue=labels, s=100, palette=colors_plot)
```

```python
sns.scatterplot(kmedoids.cluster_centers_[:, 0], kmedoids.cluster_centers_[:, 1], marker='o', color="red", label='centroids', s=100)
plt.title("K-Medoids Clustering")
plt.legend()
plt.show()
```
